## Loading Data

```python
import pandas as pd

data_source = r"C:\Users\521169\Documents\Openclassrooms\Projet 2_ Analyse des données(EdStats All indicator Query)\Data\EdStatsData.csv"
data = pd.read_csv(data_source)
data.head(5)
```

## Remove years lacking information

```python
data1 = data.drop(columns = ['1970','1971','1972','1973','1974','1975','1976','1977','1978',
                             '1979','1980','1981','1982','1983','1984','1985','1986','1987',
                             '1988','1989','1990','1991','1992','1993','1994','1995','1996',
                             '1997','1998','1999','2000','2001','2002','2003','2004','2005',
                             '2006','2007','2008','2009','2016','2017','2020','2025','2030',
                             '2035','2040','2045','2050', '2055','2060','2065','2070','2075',
                             '2080','2085','2090', '2095','2100','Unnamed: 69'
                           ])
data1
```

## Select 4 favorable indicators

```python
data2 = data1.loc[data['Indicator Name'].isin(['Internet users (per 100 people)','Population, ages 15-24, total',
                                                'GDP per capita (current US$)','Youth literacy rate, population 15-24 years, both sexes (%)'
                                               ]), 
                     :]
data2
```

## Removes country differences

```python
lignes = ['Arab World','East Asia & Pacific','East Asia & Pacific (excluding high income)','Euro area','Europe & Central Asia',
         'Europe & Central Asia (excluding high income)','European Union','Heavily indebted poor countries (HIPC)','High income',
         'Latin America & Caribbean','Latin America & Caribbean (excluding high income)','Least developed countries: UN classification',
         'Low & middle income','Low income','Lower middle income','Middle East & North Africa','Middle income','North America','OECD members'
         ,'South Asia','Sub-Saharan Africa','Sub-Saharan Africa (excluding high income)','Upper middle income','World',
         'Middle East & North Africa (excluding high income)','American Samoa','Virgin Islands (U.S.)','Central African Republic','Cayman Islands'
         ]

data3 = data2[~data1['Country Name'].isin(lignes)] # '~' permet d'obtenir les lignes inverses de la liste ligne(évênement contraire)

data3.head(10)
```

## Deletes rows with at least 3 missing values (NaN)

```python
import numpy as np

# Suppression des lignes avec des valeurs manquantes dans les colonnes numériques
data4 = data3.dropna(subset=data3.select_dtypes(include=[np.number]).columns, thresh=3)
data4
```

```python
data4.isnull().sum()
```

## Remplissage des cases avec un NAN

```python
data5 = data4.apply(lambda x: x.fillna(method='ffill') if x.dtype in [np.float64, np.int64] else x,axis = 1)
data5
```

###  Moyenne, Médiane et Ecart-type

```python
numeric_colonne = data5.select_dtypes(include='number')

# Calcul de la moyenne,médiane et ecart-type
data5['Moyenne'] = numeric_colonne.mean(axis=1)
data5['Médiane'] = numeric_colonne.median(axis=1)
data5['ecart_type'] = numeric_colonne.std(axis=1)
data6 = data5
data6
```

```python
data7 = data6.sort_values(by='Médiane', ascending = False)
data7
```

```python
data8 = data7.drop(columns=['2010','2011','2012','2013','2014','2015','Country Code','Indicator Code','Moyenne','ecart_type'])
data8 
```

```python
data8.isnull().sum()
```

```python
table = pd.pivot_table(data8, index=['Country Name'], columns=['Indicator Name'],values ='Médiane')
table
```

## Normalisation par la colonne

```python
def min_max_scaling(column):
    min_val = column.min()
    scaled_values = (column - min_val) / (column.max() - min_val)
    return scaled_values

normalisation = table.apply(min_max_scaling)
table1 = normalisation
table1
```

```python
table2 = table1.fillna(0)
table2
```

```python
table2["score"] = 4*table2["GDP per capita (current US$)"] + 5*table2["Internet users (per 100 people)"] + 3*table2["Youth literacy rate, population 15-24 years, both sexes (%)"] + 2*table2["Population, ages 15-24, total"]
table2
```

## Classement des scores par ordre décroissant

```python
table3 = table2.sort_values(by='score', ascending = False).head(10)
table3
```
